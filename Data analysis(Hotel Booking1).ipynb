```python
df.drop(['company', 'agent'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df = df[df['adr']<5000]
```

```python
df.describe()
```

```python
plt.figure(figsize=(5, 4))
plt.title('Reservation Status Count')

# Correct column reference and plot the data
plt.bar(['not canceled', 'canceled'], 
        df['is_canceled'].value_counts(), 
        edgecolor='k', 
        width=0.7)  
plt.show() 
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Adjust the legend placement
ax1.legend(bbox_to_anchor=(1, 1), title='Reservation Status')

# Add title and labels
plt.title("Reservation Status in Different Hotels", size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')

plt.show()
```

```python
# Filter rows where the hotel type is 'Resort Hotel'
resort_hotel = df[df['hotel'] == 'Resort Hotel']

# Get value counts with normalization
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# Filter rows where the hotel type is 'Resort Hotel'
city_hotel = df[df['hotel'] == 'City Hotel']

# Get value counts with normalization
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean() 
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean() 
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate: City and Resort Hotels', fontsize=30)

plt.plot(resort_hotel.index, resort_hotel['adr'], label="Resort Hotel")

plt.plot(city_hotel.index, city_hotel['adr'], label="City Hotel")

plt.legend()
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df)

```

```python
grouped_df = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="month", y="adr", data=grouped_df)

# Add titles and labels
plt.title('Total ADR for Canceled Reservations by Month')
plt.xlabel('Month')
plt.ylabel('Total ADR')
plt.show() 
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8)) 
plt.title('Top 10 country with resvation cancelled')
plt.pie(top_10_countries, autopct ='%.2f', labels = top_10_countries.index)
plt.show()  
```
